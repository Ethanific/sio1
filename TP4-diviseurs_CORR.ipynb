```python
n = int(input("Entrer un nombre entier : "))
cpt = 0
for d in range(1, n+1):
    r = n%d
    if r == 0 :
        print(d)
        cpt = cpt + 1
print("le nombre de diviseurs de", n, "vaut : ", cpt)
```

#### Exercice 3

```python
def nb_div(n):
    cpt = 0
    for d in range(1, n+1):
        r = n%d
        if r == 0 :
            cpt = cpt + 1
    return cpt
```

```python
nb_div(6)
```

```python
nb_div(7)
```

#### Exercice 4

```python
def premier(n):
    cpt = 0
    for d in range(1, n+1):
        r = n%d
        if r == 0 :
            cpt = cpt + 1
    return cpt==2
```

```python
premier(7)
```

```python
premier(6)
```

#### Exercice 5

```python
N = 100
for i in range(1,N+1):
    if premier(i):
        print(i)
```
