# Parcours d'un tableau

## Que signifie "parcourir un tableau" ?

On utilise l'expression "parcourir un tableau" lorsqu'**une variable prend successivement toutes les valeurs contenues dans un tableau**. 

Ceci peut se réaliser en itérant sur les **indices** d'un tableau

    for i in range(len(t)):
        v = t[i]

Dans ce cas, la varible `v` prend successivement toutes les valeurs du tableau `t`. 

On a donc un **parcours** du tableau.

# Exercice 1.A : indice_du_max

1. Compléter le code de la fonction `indice_du_max(t)` pour que :
  * si t est un tableau non vide, contenant des valeurs numériques
  * alors la fonction renvoie **l'indice de la plus grande valeur de t**
  * si cette valeur apparaît plusieurs fois dans t,     alors l'indice renvoyé est le plus petit possible


```
def indice_du_max(t):
    L = len(t)
    M = t[0]
    ind = 0
    for i in range(L) :
        v = t[i]
        if v > M :
            M = v
            ind = i
    return ind

print(indice_du_max([1000,50,100,30,70,80])) 
```

2. Tester votre travail avec les exemples ci-dessous

```
print(indice_du_max([10,50,100,30,70,80]) == 2)
print(indice_du_max([10,50,100,30,70,80,100]) == 2)
print(indice_du_max([10,50,100,30,70,80,3000]) == 6)
print(indice_du_max([10000,50,100,30,70,80]) == 0)
```

# Exercice 1.B : indice_du_dernier_max

1. Compléter le code de la fonction `indice_du_max(t)` pour que :
  * si t est un tableau non vide, contenant des valeurs numériques
  * alors la fonction renvoie **l'indice de la plus grande valeur de t**
  * si cette valeur apparaît plusieurs fois dans t,     alors l'indice renvoyé est **le plus grand possible**.

```
def indice_du_dernier_max(t): # CODE à MODIFIER
    L = len(t)
    M = t[0]
    ind = 0
    for i in range(L) :
        v = t[i]
        if v >= M :
            M = v
            ind = i
    return ind
```

2. Tester votre travail avec les exemples ci-dessous

```
print(indice_du_dernier_max([10,50,100,30,70,80]) == 2 )
print(indice_du_dernier_max([10,50,100,30,70,80,100]) == 6)
print( indice_du_dernier_max([10,50,100,30,70,80,3000]) == 6)
```
