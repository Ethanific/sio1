#### Exo 5 (Pour aller plus loin)

En adaptant le travail déjà réalisé pour les lettres MAJUSCULES, traiter le cas des lettres minuscules : 
    
Écrire une fonction rang(car) qui  renvoie le rang du paramètre car si c'est une minuscule, et qui renvoie – 1 sinon

```python
alphabet = 'abcdefghijklmnopqrstuvwxyz'
def rang(car) :
    resu = -1
    i = 0
    while i < len(alphabet) and resu==-1 :
        if car == alphabet[i] :
            resu = i
    return resu
```
