### Exercice 2

A l’aide de l’exercice 1, écrire une fonction `premier(n)` qui :
- prend comme paramètre un nombre entier n ≥ 2
- renvoie un booléen : 
  * True si n est premier, 
  * False sinon

* identifier les lignes de codes utiles pour programmer la fonction dans le programme de l'exercice 1
* écrire l'entête de la fonction (1ère ligne de code)
* écrire la dernière ligne de code de la fonction

```python
n = int(input('Entrer un entier :'))
resu = True
d = 2
while (d<n):
    r = n%d
    if r==0:
        resu = False
        print(n, 'est multiple de', d)
    d = d+1
if resu:
    print(n, 'est premier')
else:
    print(n, 'est composé')
```

```python
def premier(n):
    resu = True
    d = 2
    while (d<n) and resu:
        r = n%d
        if r==0:
            resu = False
            print(n, 'est multiple de', d)
        d = d+1
    return resu
```

### Exercice 4
#### question 1)
Ecrire un programme permettant d'afficher les N  plus petits nombres premiers, où N est un entier saisi par l'utilisateur.

```python
def premier(n):
    resu = True
    d = 2
    while (d<n) and resu:
        r = n%d
        if r==0:
            resu = False
        d = d+1
    return resu

N = int(input("Entrer un nombre entier : "))
print('Voici les', N, 'plus petits entiers premiers' )
cpt = 1     # le nombre premier n° 1
nombre = 2  # est égal à 2
while cpt <= N : # tant qu'on n'a pas encore atteint N nombres premiers affichés
    if premier(nombre): # on teste si la valeur nombre convient
        print(nombre)   # et si c'est le cas on l'affiche
        cpt = cpt + 1   # et on comptabilise ce nombre dans la variable cpt
    # puis on teste le nombre suivant
    nombre = nombre + 1
```

#### EX4-question 3)
En déduire un programme qui affiche uniquement le N-ième nombre premier.

Indication : il suffit de modifier le programme de la question précédente...

```python
def premier(n):
    resu = True
    d = 2
    while (d<n) and resu:
        r = n%d
        if r==0:
            resu = False
        d = d+1
    return resu


N = int(input("Entrer un nombre entier : "))
print('Voici le ', N, 'ème nombre premier' , sep='')

# le plus petit entier premier est 2
nombre = 2
cpt = 1
while cpt <= N : # tant qu'on n'a pas encore atteint N nombres premiers 
    if premier(nombre): # on teste si la valeur nombre convient
        if cpt == N : 
            print(nombre)   # on l'affiche uniquement si c'est le N-ième !
        cpt = cpt + 1   # mais on comptabilise chaque nombre premier dans la variable cpt
    # puis on passe au nombre suivant
    nombre = nombre + 1
```

#### EX4-question 4

Tester l'efficacité de votre programme pour de « grandes valeurs de N »


On peut encore améliorer la fonction premier en testant uniquement les diviseurs p qui sont inférieurs à la racine carrée de n, et donc qui ont un carré inférieur ou égal à n.

on écrit donc dans la boucle `while`: 

    while (d**2 <= n)

```python
def premier(n):
    resu = True
    d = 2
    while (d**2 <= n) and resu: # remarquer la condition de boucle
        r = n%d
        if r==0:
            resu = False
        d = d+1
    return resu


N = int(input("Entrer un nombre entier : "))
print('Voici le ', N, 'ème nombre premier' , sep='')
nombre = 2
cpt = 1
while cpt <= N : 
    if premier(nombre): 
        if cpt == N : 
            print(nombre)  
        cpt = cpt + 1   
    nombre = nombre + 1
```
