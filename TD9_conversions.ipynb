```python
def valeur(car):
    chiffre_list = '0123456789ABCDEF'
    resu = -1
    for i in range(len(chiffre_list)):
        if chiffre_list[i]==car :
            resu = i
    return resu
```

####     1) Ecrire une fonction `convert(chaine)`

* qui prend comme paramètre une chaine de caractères contenant les chiffres d'un nombre **entier en base 16**
* et qui retourne comme résultat ce nombre, converti en base 10  
* on pourra éventuellement faire AFFICHER le détail du calcul

par exemple 
* `convert('1C')` doit renvoyer **28**,  et éventuellement afficher `1 * 16**1` et `12 * 16**0`
* `convert('1A8')` doit renvoyer **424**,  et éventuellement afficher `1 * 16**2` ,  `10 * 16**1` et `8 * 16**0`

####     2) Ecrire une fonction `convert_frac(chaine)`
* qui prend comme paramètre une chaine de caractères contenant les chiffres d'un **nombre fractionnaire en base 16**
* et qui retourne comme résultat ce nombre, converti en base 10  
* on pourra éventuellement faire AFFICHER le détail du calcul

par exemple   
* `convert('10,C')` doit renvoyer 16,75
* `convert('1A,88')` doit renvoyer 26.53125
