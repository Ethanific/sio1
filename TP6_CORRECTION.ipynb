```python
def compte_lettre(car, txt):
    cpt = 0
    for i in range(len(txt)):
        lettre = txt[i]
        if lettre == car : 
            cpt = cpt+1
    return cpt
```

```python
compte_lettre('A', 'ACEBD')
```

```python
compte_lettre('I', 'ICI')
```
