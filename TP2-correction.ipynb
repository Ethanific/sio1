Pour exécuter dans Basthon, on intègre les affichages dans le message affiché par la commande `input()`

```python
from random import *

nb_essai = 4
MAX = 15
compteur=0
resultat = randint(1,MAX)
# adaptation du message pour exécution dans BASTHON
message = 'Un nombre entre 1 et ' + str(MAX) + ' est choisi : vous avez ' +str(nb_essai)+ ' essais pour le deviner\n'

gagne = False #initialement, le joueur n'a pas (encore) gagné
#la boucle continue tant que 2 conditions sont réunies
    # il reste des essais
    # et le joueur n'a pas encore gagné
while compteur < nb_essai and not gagne :
    compteur = compteur + 1
    message = message + 'Essai n°' + str(compteur) + ' : '
    n=int(input(message))
    if (n<resultat):
        message = str(n) + ' est trop petit : entrer une valeur plus grande\n'
    if (n>resultat):
        message = str(n) +  ' est trop grand: entrer une valeur plus petite\n'
    if (n == resultat):
        gagne = True
        
if gagne:
    print('Vous avez gagné au bout de', compteur, 'essais')
else :
    print('Désolé, vous avez perdu')
```
